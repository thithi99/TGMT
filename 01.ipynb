```python
#import thư viện
from PIL import Image
```

```python
import os 
print(os.getcwd()) #xem thư mục hiện tại mình đang ở đâu
#chọn thư mục chứa ảnh
dir = 'C:/Users/ADMIN/Documents/TGMT_lop1/hinhanh/minion'
```

```python
#tạo đường dẫn đến ảnh
image_path = dir + '/minion.jpg'
print(image_path)
```

```python
#sử dụng hàm Image.open để đọc ảnh
img01 = Image.open(image_path)
```

```python
#hiển thị hình ảnh
display(img01)
```

```python
#định dạng hình ảnh xem nó ở định dạng nào
print('định dạng ảnh: ', img01.format)
```

```python
#xem size
print('kích thước: ', img01.size)
```

```python
#chuyển đổi ảnh thành định dạng mới
new_img_01_path = dir + "/new_minion.jpg"
new_img_01_new_path = dir + "/new_minion.png"
img01.save(new_img_01_path)
img01.save(new_img_01_new_path)
```

```python
#đóng tập tin
img01.close()
```
