```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = 'C:/Users/ADMIN/Documents/TGMT_lop1/hinhanh/minion/minion2.jpg'

img = Image.open(my_path)
plt.imshow(img)
```

```python
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
plt.figure()
plt.hist(img_array.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
# Lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```
